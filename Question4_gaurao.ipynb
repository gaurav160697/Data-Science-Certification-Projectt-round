```python
# Calculate the attendance count and missed count for each type
df1['Attended_Count'] = df1[df1 == 'Attended'].count(axis=1)
df1['Missed_Count'] = df1[df1 == 'Missed'].count(axis=1)
df1["total_sessions"] = len(df1.columns)
df1['Attended_Percentage'] = (df1['Attended_Count'] / df1['total_sessions']) * 100
df1['Missed_Percentage'] = (df1['Missed_Count'] / df1['total_sessions']) * 100

df1.head(3)
```

```python
student= df1[df1['Type'] == 'STUDENT']
working = df1[df1['Type'] == 'WORKING']
ignore = df1[df1['Type'] == 'IGNORE']
dropout = df1[df1['Type'] == 'DROPOUT']

Types = ['student', 'working', 'ignore', 'dropout']

attended_percentage = [
    student['Attended_Percentage'].mean(),
    working['Attended_Percentage'].mean(),
    ignore['Attended_Percentage'].mean(),
    dropout['Attended_Percentage'].mean()
]

missed_percentage = [
    student['Missed_Percentage'].mean(),
    working['Missed_Percentage'].mean(),
    ignore['Missed_Percentage'].mean(),
    dropout['Missed_Percentage'].mean()
]


ignore.head()
```

```python
# Set the width of each bar
bar_width = 0.3

# Set the positions of the bars on the x-axis
r1 = range(len(Types))
r2 = [x + bar_width for x in r1]

# Create the bar plot
plt.figure(figsize=(15, 6))
plt.bar(r1, attended_percentage, color="hotpink", width=bar_width, edgecolor="black", label='Attended')
plt.bar(r2, missed_percentage, color="b", width=bar_width, edgecolor="black", label='Missed')

# Add labels, title, and legend
plt.xlabel('Type')
plt.ylabel('Percentage')
plt.title('Attendance Percentage by Type')
plt.xticks([r + bar_width/2 for r in range(len(Types))], Types)
plt.legend()
# Show the plot
plt.show()
```

```python
x1 = np.array(student['Attended_Percentage'].mean())
x2 = np.array(working['Attended_Percentage'].mean())
x3 = np.array(ignore['Attended_Percentage'].mean())
x4 = np.array(dropout['Attended_Percentage'].mean())

y = np.array([x1, x2, x3, x4])

Types = ['student', 'working', 'ignore', 'dropout']
mycolors = ["y", "hotpink", "b", "#4CAF50"]

plt.pie(y, labels = Types, colors = mycolors, autopct='%1.1f%%')
plt.title('Pie Chart of Total Attended')

plt.show() 


```

```python
#from above chart 53.7% of "Student" have attended in total sessions
#46.3% student from "Working" have attended the total sessions
#ignore and drop out occupied 0% in attendancce
```

```python
y1 = np.array((student['Missed_Percentage'].mean(),
    working['Missed_Percentage'].mean(),
    ignore['Missed_Percentage'].mean(),
    dropout['Missed_Percentage'].mean()))

Types = ['student', 'working', 'ignore', 'dropout']
mycolors = ["y", "b", "r", "#4CAF50"]

plt.pie(y1, labels = Types, colors = mycolors, autopct='%1.1f%%')
plt.title('Pie Chart of Total Missed')
plt.show() 
```

```python
#from class1 in total count 33.7% of ignore and dropout type people have missed sessions and need to get attention
# 15% of student from type "Student" missed sessions needs to get attention
#17.6% of student from "Working" missed sessions and need attention
```
